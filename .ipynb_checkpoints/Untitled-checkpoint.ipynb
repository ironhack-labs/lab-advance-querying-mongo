```python
import pymongo
import pandas as pd
import re
import json
from pandas.io.json import json_normalize


MongoClient = pymongo.MongoClient
client = MongoClient()
db = client.companies
```

```python
# Check all documents are imported
cursor = db.companies.find()
data = list(cursor)
len(data)
```

```python
# Query and convert to Dataframe
query = {'$and': [{'$or': [{'category_code': 'software'},
                 {'category_code': 'social'}, 
                 {'category_code': 'web'}]},
         {'number_of_employees': {'$gte': 50}},
         {'deadpooled_year': None},
         {'$where': 'this.funding_rounds.length >= 1'},
         {'founded_year': {'$gte': 1999}},
         {'total_money_raised': {'$regex':'\W\w*.?\w*?[MB]'}}]} 

cursor = db.companies.find(query)
df = pd.DataFrame(list(cursor))
df.head()
```

```python
#Select certain columns for further analysis.

df = df[['name', 'acquisitions', 'founded_year', 'milestones', 'number_of_employees', 'offices', 'total_money_raised', 'funding_rounds']]
df.head()
```

```python
# Three functions. Two are to flatten the data inside columns which are in json format. 
# The difference between the two if the final if statement.

def flatten_none(column, area1, area2):
    return [[i[area1], i[area2]] for e in df[column] for i in e if i[area1] != None]

def flatten(column, area1, area2):
    return [[i[area1], i[area2]] for e in df[column] for i in e]

def split(table, column):
    length = [len(e) for e in df[column]]
    data = []
    x = 0
    for i in range(0, len(table)): 
        if x < 204:
            data.append(table[i: i + length[x]])
        x += 1
    return data
```

```python
# Classify the companies into big corporations and otherwise. I have done this by how much 
# money they have raised.

clas = [[' /Big Corp'] if re.match('\W\w*.?\w*?B', elem) is not None else [' /Otherwise']for elem in df['total_money_raised']] 
clas = pd.DataFrame(clas)
clas.head()
```

```python
# Apply the split function and merge two columns into one to obtain the dataframe below
# showing the corrdinates of the various offices of each company.
coord = flatten_none('offices', 'longitude', 'latitude') 
coords= pd.DataFrame(split(coord, 'offices'))
coords['Company'] = df['name']
coords['Class'] = clas[0]
coords['Company/ Class'] = coords['Company'].astype(str) + coords['Class']
coords = coords.drop(columns = ['Company', 'Class'])
coords.head()
```

```python
# Function to tranform the dataframe above into the format below.

def transform(database, column):
    x = 0
    y = 0
    l = []
    length = [len(e) for e in df[column]]
    for i in database['Company/ Class']:
        for e in range(length[x]):
            l.append([i + str(y), coords[e][x]])
            y += 1
        x += 1
        y = 0
    return l

colnames = ['Company/ Class', 'Coordinates']    
coords = pd.DataFrame(transform(coords, 'offices'), columns = colnames)
coords = coords.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
coords.head()
```

```python
# Seperate latitude and longitude into different columns for later stage.
a = [e for e in coords['Coordinates']]
colnames = ['longitude', 'latitude']
lat_long = pd.DataFrame(a, columns = colnames)
lat_long['Company/ Class'] = coords['Company/ Class']
lat_long.head()
```

```python
# Function to format the coordinates in a JSON format.
def toGeoJSON(array):
    return {
        "type":"Point",
        "coordinates":[array[0],array[1]]
    }

coords['Coordinates'] = coords.apply(lambda r: toGeoJSON(r['Coordinates']), axis=1)
coords.head()
```

```python
# Export dataframe as a JSON file.
coords.to_json('visualize_companies.json', orient="records",lines=True)
```
