### Lab Advance Query Pymongo

```python
#Tras iniciar servidor de DB mongo, importo módulo 'pymongo'.
import pymongo
#Asigno variable al cliente de DB y a la DB.
MongoClient = pymongo.MongoClient
client = MongoClient()
db = client.companies
#Verifico que se han importado todos los documentos.
cursor = db.companies.find()
data = list(cursor)
print(len(data))
```

1. All the companies that it's name match 'Babelgum'. Retrieve only their name field.

```python
query1 = db.companies.find({'name':"Babelgum"})
print(list(map(lambda r:r['name'],list(query1))))
```

2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by number of employees.

```python
query2 = db.companies.find({'number_of_employees':{'$gte':5000}})\
.sort('number_of_employees',pymongo.ASCENDING).limit(20)
print(list(map(lambda r:r,list(query2))))
```

3. All the companies founded between 2000 and 2005, both years included. Retrieve only the name and founded_year fileds.

```python
query3 = db.companies.find({'founded_year':{'$gte':2000,'$lte':2005}})
print(list(map(lambda r:[(r['name']),(r['founded_year'])],list(query3))))
```

4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the name and ipo fields.

```python
query4 = db.companies.find({'ipo.valuation_amount':{'$gte':100000000},'$and':[{'founded_year':{'$lte':2010}}]})
print(list(map(lambda r:[(r['name']),(r['ipo'])],list(query4))))
```

5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
query5 = db.companies.find({'number_of_employees':{'$lte':1000},'$and':[{'founded_year':{'$lte':2005}}]})\
.sort('number_of_employees',pymongo.ASCENDING).limit(10)
print(list(map(lambda r:r,list(query5))))
```
